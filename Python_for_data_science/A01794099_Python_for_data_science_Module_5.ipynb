##Numpy 1D Arrays

**Numpy** is a library for scientific computing.  Numpy is also the basis for pandas.

A "numpy" array or "ndarray" is similar to a list. It's usually fixed in size and each element is of the same type, in this case, integers. We can cast a list to a numpy array by first importing numpy. We then cast the list as follows. We can access the data via an index. As with a list, we can access each element with an integer and a square bracket.

```python
#ndarray

import numpy as np
a=np.array([0,1,2,3,4])
print(a.size)
print(a.ndim) #dimension
print(a.shape)
```

```python
b=np.array([3.1,11.02,6.2,213.2,5.2])
type(b)
```

```python
#dot product
u=np.array([1,2])
v=np.array([3,1])
result=np.dot(u,v)
result
```

```python
#Broadcasting
u=np.array([1,2,3,-1])
z=u+1
z
```

```python
#Universal functions
a=np.array([1,-1,1,-1])
mean_a=a.mean()
mean_a
```

```python
b=np.array([1,-2,3,4,5])
max_b=b.max()
max_b
```

##Numpy 2D Arrays

```python
import numpy as np
A=np.array([[11,12,13],[21,22,23],[31,32,33]])
print(A.ndim)
print(A.shape)
print(A.size)
```

```python
#Add arrays
X=np.array([[1,0],[0,1]])
Y=np.array([[2,1],[1,2]])
Z=X+Y
Z
```

```python
#mMultiplying array
Y=np.array([[2,1],[1,2]])
Z=2*Y
Z
```

```python
#PRODUCT
X=np.array([[1,0],[0,1]])
Y=np.array([[2,1],[1,2]])
Z=X*Y
Z
```

```python
#Matrix multiplication
A=np.array([[0,1,1],[1,0,1]])
B=np.array([[1,1],[1,1],[-1,1]])
C=np.dot(A,B)
C
```

##Simple APIs

**Application Program Interface (API)**

An API lets two pieces of software talk to each other.

You use the API to communicate with the API via inputs and outputs.

We use the pandas API to process the data by communicating with the other Software Components.

REST APIs are another popular type of API; they allow you to communicate through the internet allowing you to take advantage of resources like storage, access more data, artificial intelligent algorithms, and much more.

The RE stands for Representational, the S stands for State, the T stand for Transfer. In rest API’s your program is called the client.


HTTP methods are a way of transmitting data over the internet.

The request is usually communicated via an HTTP message.
The HTTP message usually contains a JSON file. This contains instructions for what operation we would like the service to perform. This operation is transmitted to the webservice via the internet. The service performs the operation.
In the similar manner, the webservice returns a response via an HTTP message, where the information is usually returned via a JSON file.

image.png

An API key is a way to access the API. It’s a unique set of characters that the
API uses to identify you and authorize you. Usually your first call to the API includes the API key. This will allow you access to the API.
